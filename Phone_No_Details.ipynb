```python
import phonenumbers
from phonenumbers import geocoder, carrier, timezone

def track_phone_number(phone_number):
    try:
        # Parse the phone number with region set to 'IN' (India)
        parsed_number = phonenumbers.parse(phone_number, "IN")

        # Validate the phone number
        if not phonenumbers.is_valid_number(parsed_number):
            return "Invalid phone number."

        # Get details about the phone number
        location = geocoder.description_for_number(parsed_number, "en")  # Location (circle)
        operator = carrier.name_for_number(parsed_number, "en")          # Telecom operator
        timezones = timezone.time_zones_for_number(parsed_number)        # Time zones

        # Construct the result
        result = {
            "Phone Number": phone_number,
            "Location": location,
            "Operator": operator,
            "Time Zones": list(timezones)
        }
        return result

    except phonenumbers.NumberParseException as e:
        return f"Error: {e}"

# Example usage
if __name__ == "__main__":
    phone_number = input("Enter the phone number (with country code): ")
    result = track_phone_number(phone_number)
    print(result)
```
